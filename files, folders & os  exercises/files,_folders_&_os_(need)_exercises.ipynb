```python
#Task 1

#os-collaboration-exercise-data
#We like to convert the filename to a more useful numerical format. 
#Specifically, we want to convert month-date-city.txt. (i.e. may-10-singapore.txt) to a mm-dd-city.txt (i.e. 05-10-singapore.txt) format.

#Using the following dictionary (or otherwise), write a snippet of Python code to convert 'oct-08_singapore.txt' to 10-08-singapore.txt.
import os
import glob
import shutil

month_dict = {'jan': '01', 'feb': '02', 'mar': '03',
             'apr': '04', 'may': '05', 'jun': '06',
             'jul': '07', 'aug': '08', 'sep': '09',
             'oct': '10', 'nov': '11', 'dec': '12'
}

#os.chdir(r'C:\Users\ediso\Documents\GitHub\learning-portfolio-Herodotus9\files, folders & os  exercises\os-collaboration-exercise-data')
#all_file = '\ediso\Documents\GitHub\learning-portfolio-Herodotus9\files, folders & os  exercises\os-collaboration-exercise-data'
all_files = glob.glob('os-collab*/*')


try:
    for filepath in all_files:
        direct, filename = os.path.split(filepath)  #splitting into individual components
        title, ext = os.path.splitext(filename)   
        month, daycity = title.split('-')
        date, city = daycity.split('_')
    
        new_month = month_dict[month]  #referring to dicitionary to change month into numbers
        
        new_filename = f"{new_month}-{date}-{city}{ext}"
        new_filepath = os.path.join(direct, new_filename)
        
        os.rename(filepath, new_filepath)
        print('files have been renamed!')

except:
    print('Files has already been changed!')
    
```

```python
#Task 5
#Use a for loop to create a folder for each city.
#The list of cities is provided below for your convenience.

import os
import glob
import shutil

city_list = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
             'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

dire = f"os-collaboration-exercise-data"

for city in city_list:
    city_folder = os.path.join(dire, city)
    
    if os.path.exists(city_folder):
        print(f'{city_folder} already exists; skipping creation.')
    
    else:
        print(f'Creating {city_folder}')
        os.mkdir(city_folder)
```

```python
#Task 6
#Use glob to list all the files from Cairo.
#Now use a for loop and shutil.copy() to copy all the files related to Cairo to the corresponding folder you created.

import os
import glob
import shutil

dire = f"os-collaboration-exercise-data"

cairo_all = glob.glob('os-collab*/*cairo*')

for cairo_file in cairo_all:
    path_to_destination = os.path.join(dire,'Cairo', os.path.basename(cairo_file))
    try:
        if os.path.exists(path_to_destination):
            print(f'{os.path.basename(cairo_file)} has already been copied to folder: Cairo')
        else:
            shutil.copy(cairo_file, path_to_destination)
            print(f'Copied {os.path.basename(cairo_file)} to Cairo')
    except PermissionError:
        continue  #XD i made the code ignore the permission error that arises from the code trying to copy the open file "Cairo" into itself
```
