```python
import os
import glob
import shutil

def city_move(city_name):  
    dire = f"os-collaboration-exercise-data"
    
    city_all = glob.glob(f"os-collab*/*{city_name.lower()}*")  
    total_files_moved = 0

    for index, city_file in enumerate(city_all, start=1):
        path_to_destination = os.path.join(dire, str(city_name), os.path.basename(city_file))

        if os.path.exists(path_to_destination):
            #print(f'{os.path.basename(city_file)} has an identical file inside the folder: {city_name}!')
            count.append(city_file)
        elif os.path.basename(city_file) == city_name:
            continue
        else:
            #shutil.copy(city_file, path_to_destination)
            shutil.move(city_file, path_to_destination)
            total_files_moved += 1
            if total_files_moved % 100 == 0:
                print(f"{total_files_moved} files moved...")
            #print(f'Moved {os.path.basename(city_file)} to {city_name} Folder')

    print(f"Total files moved for {city_name}: {total_files_moved}")
    return total_files_moved

city_list = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila', 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
count = []
total_files_moved_all = 0
for city_name in city_list:
    total_files_moved_all += city_move(city_name)

files_outside = len(count)
if files_outside > 0:
    ask = input(f"There are {files_outside} identical files outside the folders.\nDo you wish to delete them? Type Y if you wish to delete files.")
    if ask.upper() == "Y":
        for planned_deletion in count:
            if os.path.exists(planned_deletion):  # Check if the file exists before attempting to delete
                os.remove(planned_deletion)
                print(f"File '{planned_deletion}' has been deleted")
        print("All files have been deleted.")
    else:
        print("Understood, no files will be deleted.")
else:
    print("There are no identical files outside the folders.")
print(f"Total files moved across all cities: {total_files_moved_all}")
#1
```
